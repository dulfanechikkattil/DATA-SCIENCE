```python
# Create Sentiment Labels for sentimental analysis
def sentiment_label(rating):
    if rating >= 4:
        return 1
    elif rating == 3:
        return 0
    else:
        return -1

data['sentiment'] = data['overall'].apply(sentiment_label)
```

# Text Preprocessing (Cleaning, Stopwords, Lemmatization)

```python
# 1️⃣ Import required libraries
```

```python
import string
import nltk

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# 2️⃣ Create preprocessing objects
```

```python
punctuation = string.punctuation
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
# 3️⃣ Text cleaning function 
```

```python
def clean(text):
    words = word_tokenize(text.lower())   # lowercase + tokenize
    
    words = [word for word in words 
             if word not in stop_words and word not in punctuation]
    
    words = [lemmatizer.lemmatize(word) for word in words]
    
    return " ".join(words)
```

```python
# 4️⃣ Apply cleaning to entire dataset
```

```python
data['clean_review'] = data['reviewText'].apply(clean)
```

```python
data
```

# Feature Extraction (CountVectorizer / TF-IDF)

```python
x=data['clean_review']
x
```

```python
y=data['sentiment']
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

x_train = cv.fit_transform(x_train)
x_test = cv.transform(x_test)
```

# Model Building

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr

lr=lr.fit(x_train,y_train)
y_pred_lr = lr.predict(x_test)
```

from sklearn.metrics import accuracy_score
acc_lr = accuracy_score(y_test, y_pred_lr)
print("Accuracy:", acc_lr)

**RandomForestClassifier**

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200)
rf
rf.fit(x_train, y_train)

y_pred_rf= rf.predict(x_test)
```

```python
acc_rf = accuracy_score(y_test, y_pred_rf)
print("Accuracy:", acc_rf)
```

**MultinomialNB**

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb.fit(x_train, y_train)
y_pred_nb = nb.predict(x_test)
```

```python
acc_nb=accuracy_score(y_test, y_pred_nb)
print("Accuracy:", acc_nb)
```

**SVM**

```python
from sklearn.svm import LinearSVC

svm = LinearSVC()
svm.fit(x_train, y_train)
y_pred_svm = svm.predict(x_test)
```

```python
acc_svm=accuracy_score(y_test, y_pred_svm)
print("Accuracy:", acc_svm)
```

# Model Accuracy Comparison

```python
from tabulate import tabulate
table=[['Model','Accuracy score'],['LogisticRegression',acc_lr],['RandomForestClassifier',acc_rf],
       ['MultinomialNB',acc_nb],['svm (LinearSVC)]',acc_svm]]
       
print(tabulate(table,headers='firstrow',tablefmt='fancy_grid'))
```
